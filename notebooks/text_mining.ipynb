```python
#Add spaces before and after punctuation
def create_spaces_punctuation(corpus):
  """Receives a string corpus and adds spaces before and 
  after punctuation signs"""
  corpus = corpus.replace(".", " . ")\
                 .replace(",", " , ")\
                 .replace(";", " ; ")\
                 .replace(":", " : ")\
                 .replace("!", " ! ")\
                 .replace("?", " ? ")\
                 .replace("<", " < ")\
                 .replace(">", " > ")\
                 .replace("&", " & ")\
                 .replace("(", " ( ")\
                 .replace(")", " ) ")\
                 .replace("[", " [ ")\
                 .replace("]", " ] ")\
                 .replace("#", " # ")\
                 .replace("{", " { ")\
                 .replace("}", " } ")\
                 .replace("'", " ' ")\
                 .replace("\"", " \" ")
  return corpus


def create_tokens_dicts(corpus):
  """Creates dicts with tokens as key and count as value"""
  #Create iterable list from string
  corpus = [corpus]

  #Create dictionaries 
  Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams, Sevengrams = {}, {}, {}, {}, {}, {}, {}
  list_all = [Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams, Sevengrams]

  for i in range(len(list_all)): 
    vectorizer = CountVectorizer(analyzer='word', ngram_range=(i+1, i+1), token_pattern=r"(?u)\b\w\w+\b|!|\?|\"|\'|\.|\,|\;|\:|\<|\>|\&|\(|\)|\[|\]|\{|\}|\#")
    X = vectorizer.fit_transform(corpus)
    W = vectorizer.get_feature_names_out()

    for j in range(len(W)):
      list_all[(i)][W[j]] = X.toarray()[0][j]
  
  return Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams, Sevengrams
```

```python
#REVER OS OMEGA
def cohesion_metrics_bigrams(Unigrams, Bigrams):
  #Create new dictinary that includes for each bigram the count, the scp, dice, im.
  Bigrams_metrics = {}
  N = len(Unigrams) #total number of bigrams in dictionary

  for bigram in Bigrams:
    words = bigram.split() #split the bigram into two separate words
    freq = Bigrams[bigram]
    
    #Get the first and last unigram number of occurences in the text
    first_word_count = Unigrams[words[0]]
    last_word_count = Unigrams[words[-1]]

    #Calculate the cohesion metrics 
    scp = freq ** 2 / (first_word_count * last_word_count) 
    dice = freq * 2 / (first_word_count + last_word_count) 
    mi = log(freq / (first_word_count * last_word_count)) 
    omega2 = (N * freq - first_word_count * last_word_count) ** 2 / ((first_word_count * last_word_count) * (N - first_word_count) * (N - last_word_count)) 
    
    #Dict with metrics:
    Bigrams_metrics[bigram] = [Bigrams[bigram]]

    #Add metrics to the dictionary
    metrics = [scp, dice, mi, omega2]
    for i in range(len(metrics)):
      Bigrams_metrics[bigram].append(metrics[i])

  return Bigrams_metrics


def cohesion_metrics_threegrams(Unigrams, Bigrams, Threegrams):
  #Create new dictinary that includes for each threegram the count, the scp, dice, im.
  Threegrams_metrics = {}
  N = len(Threegrams) #total number in dictionary
  n = 3 #number of words in a threegram

  for threegram in Threegrams:
    words = threegram.split() #split the bigram into two separate words
    freq = Threegrams[threegram]

    #Get the first and last unigram number of occurences in the text
    first_uni_count = Unigrams[words[0]] #first unigram inside this threegram
    last_uni_count = Unigrams[words[-1]] #last unigram inside this threegram
    #Get the first and last bigram number of occurences in the text
    first_bi_count = Bigrams[words[0] + " " + words[1]] #first bigram inside this threegram
    last_bi_count = Bigrams[words[-2] + " " + words[-1]] #last bigram inside this threegram

    #Calculate the cohesion metrics
    scp = freq ** 2 / ((1/(n-1)) * (first_uni_count *  last_bi_count + first_bi_count  * last_uni_count)) 
    dice = freq * 2 / ((1/(n-1)) * (first_uni_count +  last_bi_count + first_bi_count  + last_uni_count))  
    mi = log(freq / ((1/(n-1)) * (first_uni_count *  last_bi_count + first_bi_count  * last_uni_count))) 
    avq = (1/(n-1)) * (first_uni_count *  last_bi_count + first_bi_count  * last_uni_count)
    avd = (1/(n-1)) * ((first_uni_count *  last_bi_count) * (N - first_uni_count) * (N - last_bi_count)\
                    + (first_bi_count  * last_uni_count) * (N -first_bi_count)  * (N -last_uni_count) )
    omega2 = (N * freq - avq ** 2) / avd

    #Dict with metrics:
    Threegrams_metrics[threegram] = [Threegrams[threegram]]

    #Add cohesion metrics to the new dictionary  
    Threegrams_metrics[threegram].append(scp)
    Threegrams_metrics[threegram].append(dice)
    Threegrams_metrics[threegram].append(mi)
    Threegrams_metrics[threegram].append(omega2)

  return Threegrams_metrics


def cohesion_metrics_fourgrams(Unigrams, Bigrams, Threegrams, Fourgrams):
  #Create new dictinary that includes for each fourgram the count, the scp, dice, im.
  Fourgrams_metrics = {}
  N = len(Fourgrams) #total number in dictionary
  n = 4 #number of words in a fourgram

  for fourgram in Fourgrams:
    words = fourgram.split() #split the bigram into two separate words
    freq = Fourgrams[fourgram]

    #Get the first and last unigram number of occurences in the text
    first_uni_count = Unigrams[words[0]] #first unigram inside this fourgram
    last_uni_count = Unigrams[words[-1]] #last unigram inside this fourgram
    #Get the first and last bigram number of occurences in the text
    first_bi_count = Bigrams[words[0] + " " + words[1]] #first bigram inside this fourgram
    last_bi_count = Bigrams[words[-2] + " " + words[-1]] #last bigram inside this fourgram
    #Get the first and last threegram number of occurences in the text
    first_three_count = Threegrams[words[0] + " " + words[1] + " " + words[2]] #first threegram inside this fourgram
    last_three_count = Threegrams[words[-3] + " " + words[-2] + " " + words[-1]] #last threegram inside this fourgram

    #Calculate the cohesion metrics
    scp = freq ** 2 / ((1/(n-1)) * (first_uni_count *  last_three_count + first_bi_count * last_bi_count + first_three_count * last_uni_count))
    dice = freq * 2 / ((1/(n-1)) * (first_uni_count +  last_three_count + first_bi_count  + last_bi_count + first_three_count + last_uni_count))  #CERTO
    mi = log(freq / ((1/(n-1)) *  (first_uni_count *  last_three_count + first_bi_count * last_bi_count + first_three_count * last_uni_count)))
    avq = (1/(n-1)) * (first_uni_count *  last_three_count + first_bi_count * last_bi_count + first_three_count * last_uni_count)
    avd = (1/(n-1)) * ((first_uni_count *  last_three_count)*(N - first_uni_count)*(N - last_three_count) 
                        + first_bi_count * last_bi_count + first_three_count * last_uni_count)
    omega2 = (N * freq - avq ** 2) / avd

    #Dict with metrics:
    Fourgrams_metrics[fourgram] = [Fourgrams[fourgram]]

    #Add cohesion metrics to the new dictionary 
    Fourgrams_metrics[fourgram].append(scp)
    Fourgrams_metrics[fourgram].append(dice)
    Fourgrams_metrics[fourgram].append(mi)
    Fourgrams_metrics[fourgram].append(omega2)

  return Fourgrams_metrics


def cohesion_metrics_fivegrams(Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams):
  #Create new dictinary that includes for each fourgram the count, the scp, dice, im.
  Fivegrams_metrics = {}
  N = len(Fivegrams) #total number in dictionary
  n = 5 #number of words

  for fivegram in Fivegrams:
    words = fivegram.split() #split the bigram into two separate words
    freq = Fivegrams[fivegram]
    n = 5 #number of words

    #Get the first and last unigram number of occurences in the text
    first_uni_count = Unigrams[words[0]] #first unigram inside this fourgram
    last_uni_count = Unigrams[words[-1]] #last unigram inside this fourgram
    #Get the first and last bigram number of occurences in the text
    first_bi_count = Bigrams[words[0] + " " + words[1]] #first bigram inside this fourgram
    last_bi_count = Bigrams[words[-2] + " " + words[-1]] #last bigram inside this fourgram
    #Get the first and last threegram number of occurences in the text
    first_three_count = Threegrams[words[0] + " " + words[1] + " " + words[2]] #first threegram inside this fourgram
    last_three_count = Threegrams[words[-3] + " " + words[-2] + " " + words[-1]] #last threegram inside this fourgram
    #Get the first and last fourgram number of occurences in the text
    first_four_count = Fourgrams[words[0] + " " + words[1] + " " + words[2] + " " + words[3]] 
    last_four_count = Fourgrams[words[-4] + " " + words[-3] + " " + words[-2] + " " + words[-1]] 

    #Calculate the cohesion metrics
    scp = freq * 2 / ((1/(n-1)) * (first_uni_count * last_four_count + first_bi_count * last_three_count + first_three_count * last_bi_count + first_four_count * last_uni_count))
    dice = freq * 2 / ((1/(n-1)) * (first_uni_count +  last_four_count + first_bi_count  + last_three_count + first_three_count + last_bi_count + first_four_count + last_uni_count)) #CERTO
    mi = log(freq / ((1/(n-1)) * (first_uni_count * last_four_count + first_bi_count * last_three_count + first_three_count * last_bi_count + first_four_count * last_uni_count)))
    avq = (1/(n-1)) * (first_uni_count * last_four_count + first_bi_count * last_three_count + first_three_count * last_bi_count + first_four_count * last_uni_count)
    avd = (1/(n-1)) * ((first_uni_count * last_four_count) * (N - first_uni_count) * (N - last_four_count) + 
                       (first_bi_count * last_three_count) * (N - first_bi_count) * (N - last_three_count) + 
                       (first_three_count * last_bi_count) * (N - first_three_count) * (N - last_bi_count) + 
                       (first_four_count * last_uni_count) * (N - first_four_count) * (N - last_uni_count))
    omega2 = (N * freq - avq ** 2) / avd

    #Dict with metrics:
    Fivegrams_metrics[fivegram] = [Fivegrams[fivegram]]

    #Add cohesion metrics to the new dictionary 
    Fivegrams_metrics[fivegram].append(scp)
    Fivegrams_metrics[fivegram].append(dice)
    Fivegrams_metrics[fivegram].append(mi)
    Fivegrams_metrics[fivegram].append(omega2)

  return Fivegrams_metrics


def cohesion_metrics_sixgrams(Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams):
  #Create new dictinary that includes for each fourgram the count, the scp, dice, im.
  Sixgrams_metrics = {}
  N = len(Sixgrams) #total number in dictionary
  n = 6 #number of words in a fourgram

  for sixgram in Sixgrams:
    words = sixgram.split() #split the bigram into two separate words
    freq = Sixgrams[sixgram]

    #Get the first and last unigram number of occurences in the text
    first_uni_count = Unigrams[words[0]] #first unigram inside this fourgram
    last_uni_count = Unigrams[words[-1]] #last unigram inside this fourgram
    #Get the first and last bigram number of occurences in the text
    first_bi_count = Bigrams[words[0] + " " + words[1]] #first bigram inside this fourgram
    last_bi_count = Bigrams[words[-2] + " " + words[-1]] #last bigram inside this fourgram
    #Get the first and last threegram number of occurences in the text
    first_three_count = Threegrams[words[0] + " " + words[1] + " " + words[2]] #first threegram inside this fourgram
    last_three_count = Threegrams[words[-3] + " " + words[-2] + " " + words[-1]] #last threegram inside this fourgram
    #Get the first and last fourgram number of occurences in the text
    first_four_count = Fourgrams[words[0] + " " + words[1] + " " + words[2] + " " + words[3]] 
    last_four_count = Fourgrams[words[-4] + " " + words[-3] + " " + words[-2] + " " + words[-1]] 
    #Get the first and last fivegram number of occurences in the text
    first_five_count = Fivegrams[words[0] + " " + words[1] + " " + words[2] + " " + words[3] + " " + words[4]] 
    last_five_count = Fivegrams[words[-5] + " " + words[-4] + " " + words[-3] + " " + words[-2] + " " + words[-1]]

    #Calculate the cohesion metrics
    scp = freq * 2 / ((1/(n-1)) * (first_uni_count *  last_five_count + first_bi_count * last_four_count + first_three_count * last_three_count \
                                  + first_four_count * last_bi_count + first_five_count * last_uni_count))
    dice = freq * 2 / ((1/(n-1)) * (first_uni_count +  last_five_count + first_bi_count  + last_four_count + first_three_count + last_three_count \
                                  + first_four_count + last_bi_count + first_five_count + last_uni_count)) #CERTO
    mi = log(freq / ((1/(n-1)) * (first_uni_count *  last_five_count + first_bi_count * last_four_count + first_three_count * last_three_count \
                                  + first_four_count * last_bi_count + first_five_count * last_uni_count)) )
    avq = (1/(n-1)) * (first_uni_count *  last_five_count + first_bi_count * last_four_count + first_three_count * last_three_count \
                                  + first_four_count * last_bi_count + first_five_count * last_uni_count)
    avd = (1/(n-1)) * (first_uni_count *  last_five_count * (N - first_uni_count) * (N - last_five_count )
                        + first_bi_count * last_four_count * (N - first_bi_count) * (N - last_four_count ) 
                        + first_three_count * last_three_count * (N - first_three_count) * (N - last_three_count ) 
                        + first_four_count * last_bi_count * (N - first_four_count) * (N - last_bi_count  )
                        + first_five_count * last_uni_count * (N - first_five_count) * (N - last_uni_count ))
    omega2 = (N * freq - avq ** 2) / avd

    #Dict with metrics:
    Sixgrams_metrics[sixgram] = [Sixgrams[sixgram]]

    #Add cohesion metrics to the new dictionary 
    Sixgrams_metrics[sixgram].append(scp)
    Sixgrams_metrics[sixgram].append(dice)
    Sixgrams_metrics[sixgram].append(mi)
    Sixgrams_metrics[sixgram].append(omega2)

  return Sixgrams_metrics


def cohesion_metrics_sevengrams(Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams, Sevengrams):
  #Create new dictinary that includes for each fourgram the count, the scp, dice, im.
  Sevengrams_metrics = {}
  N = len(Sevengrams) 
  n = 7 

  for sevengram in Sevengrams:
    words = sevengram.split() #split the bigram into two separate words
    freq = Sevengrams[sevengram]


    #Get the first and last unigram number of occurences in the text
    first_uni_count = Unigrams[words[0]] #first unigram inside this fourgram
    last_uni_count = Unigrams[words[-1]] #last unigram inside this fourgram
    #Get the first and last bigram number of occurences in the text
    first_bi_count = Bigrams[words[0] + " " + words[1]] #first bigram inside this fourgram
    last_bi_count = Bigrams[words[-2] + " " + words[-1]] #last bigram inside this fourgram
    #Get the first and last threegram number of occurences in the text
    first_three_count = Threegrams[words[0] + " " + words[1] + " " + words[2]] #first threegram inside this fourgram
    last_three_count = Threegrams[words[-3] + " " + words[-2] + " " + words[-1]] #last threegram inside this fourgram
    #Get the first and last fourgram number of occurences in the text
    first_four_count = Fourgrams[words[0] + " " + words[1] + " " + words[2] + " " + words[3]] 
    last_four_count = Fourgrams[words[-4] + " " + words[-3] + " " + words[-2] + " " + words[-1]] 
    #Get the first and last fivegram number of occurences in the text
    first_five_count = Fivegrams[words[0] + " " + words[1] + " " + words[2] + " " + words[3] + " " + words[4]] 
    last_five_count = Fivegrams[words[-5] + " " + words[-4] + " " + words[-3] + " " + words[-2] + " " + words[-1]]
    #Get the first and last sixgram number of occurences in the text
    first_six_count = Sixgrams[words[0] + " " + words[1] + " " + words[2] + " " + words[3] + " " + words[4] + " " + words[5]] 
    last_six_count = Sixgrams[words[-6] + " " + words[-5] + " " + words[-4] + " " + words[-3] + " " + words[-2] + " " + words[-1]]

    #Calculate the cohesion metrics
    scp = freq * 2 / ((1/(n-1))* (first_uni_count * last_six_count + first_bi_count * last_five_count + first_three_count * last_four_count \
                                  + first_four_count * last_three_count + first_five_count * last_bi_count + first_six_count * last_uni_count))
    dice = freq * 2 / ((1/(n-1)) * (first_uni_count +  last_six_count + first_bi_count  + last_five_count + first_three_count + last_four_count \
                                  + first_four_count + last_three_count + first_five_count + last_bi_count + first_six_count + last_uni_count)) #CERTO
    mi = log(freq / ((1/(n-1)) * (first_uni_count * last_six_count + first_bi_count * last_five_count + first_three_count * last_four_count \
                                  + first_four_count * last_three_count + first_five_count * last_bi_count + first_six_count * last_uni_count)))
    avq = (1/(n-1)) * (first_uni_count * last_six_count + first_bi_count * last_five_count + first_three_count * last_four_count \
                                  + first_four_count * last_three_count + first_five_count * last_bi_count + first_six_count * last_uni_count)
    avd = (1/(n-1)) * (first_uni_count * last_six_count * (N - first_uni_count) * (N - last_six_count)
                       + first_bi_count * last_five_count * (N - first_bi_count) * (N - last_five_count) 
                       + first_three_count * last_four_count * (N - first_three_count) * (N - last_four_count)
                       + first_four_count * last_three_count * (N - first_four_count) * (N - last_three_count)
                       + first_five_count * last_bi_count * (N - first_five_count) * (N - last_bi_count)
                       + first_six_count * last_uni_count * (N - first_six_count) * (N - last_uni_count))
    omega2 = (N * freq - avq ** 2) / avd

    #Dict with metrics:
    Sevengrams_metrics[sevengram] = [Sevengrams[sevengram]]

    #Add cohesion metrics to the new dictionary 
    Sevengrams_metrics[sevengram].append(scp)
    Sevengrams_metrics[sevengram].append(dice)
    Sevengrams_metrics[sevengram].append(mi)
    Sevengrams_metrics[sevengram].append(omega2)

  return Sevengrams_metrics
```

```python
def mean_cohesion(dictionary, cohesion_code):
  """Calculates the mean value of a cohesion metrics, 
  where dictionary must be a dict with the metrics and
  cohesion_code is the position of that metric in the dict value"""
  count = 0
  sum = 0 

  for key in list(dictionary):
    count += 1
    sum += int(dictionary[key][int(cohesion_code)])
  mean = sum / count
  return mean


def min_cohesion(dictionary, cohesion_code, min):
  """Retrieves dictionaries with a given cohesion metric
  minimum value for all elements, 
  where dictionary must be a dict with the metrics,
  cohesion_code is the position of that metric in the dict value
  and min is the minimum value accepted for that metric"""

  for key in list(dictionary):
    if (dictionary[key][2]) < min: 
      del dictionary[key]
  return dictionary


def mean_for_all(dictionary):
  """Retrieves mean of all metrics at the same time for a given dictionary """
  mean_scp = mean_cohesion(dictionary, 1)
  mean_dice = mean_cohesion(dictionary, 2)
  mean_mi = mean_cohesion(dictionary, 3)
  mean_omega = mean_cohesion(dictionary, 4)
  return mean_scp, mean_dice, mean_mi, mean_omega


def min_cohesion_all(dictionary, min_scp, min_dice, min_mi, min_omega):
  """Retrieves dictionary where a minimum for all the cohesion metrics was given"""
  dictionary = min_cohesion(dictionary, 1, min_scp)
  dictionary = min_cohesion(dictionary, 2, min_dice)
  dictionary = min_cohesion(dictionary, 3, min_mi)
  dictionary = min_cohesion(dictionary, 4, min_omega)
  return dictionary
```

```python
def evaluate_extraction(retrieved_expressions, relevant_expressions):
  #Calculate true pos, false pos, true neg and false neg
  true_pos = 0
  false_pos = 0
  true_neg = 0
  false_neg = 0

  #True and false positives
  for i in range(len(retrieved_expressions)):
    if retrieved_expressions[i] in relevant_expressions:
      true_pos += 1
    else:
      false_pos += 1

  #False negatives
  for k in range(len(relevant_expressions)):
     if relevant_expressions[k] not in retrieved_expressions:
       false_neg += 1

  #True negatives
  #Calculate the sum of all the expressions from the text before we use the extractor
  all_expressions = 0
  for i in range(len(list_all)):
    all_expressions += len(list_all[i])

  true_neg = all_expressions - (false_neg + true_pos + false_pos )


  #Calculate metrics: Recall, Precision and F1-score
  recall = true_pos / (true_pos + false_neg)
  precision = true_pos / (true_pos + false_pos) 
  f1_score = 2 * true_pos / (2 * true_pos + false_neg + false_pos)

  return recall, precision, f1_score

recall, precision, f1_score = evaluate_extraction(retrieved_expressions, relevant_expressions)

print("Recall:", recall)
print("Precision:", precision)
print("F1 score:", f1_score)
```

### **Complete extractor**

```python
#Define complete extractor 
def extractor(corpus):
  #Create spaces before and after punctuation
  corpus = create_spaces_punctuation(corpus)
  #Create dictionaries with tokens
  Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams, Sevengrams = create_tokens_dicts(corpus)
  list_all = [Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams, Sevengrams]

  #Add metrics to the dictionaries
  Bigrams_metrics = cohesion_metrics_bigrams(Unigrams, Bigrams)
  Threegrams_metrics = cohesion_metrics_threegrams(Unigrams, Bigrams, Threegrams)
  Fourgrams_metrics = cohesion_metrics_fourgrams(Unigrams, Bigrams, Threegrams, Fourgrams)
  Fivegrams_metrics = cohesion_metrics_fivegrams(Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams)
  Sixgrams_metrics = cohesion_metrics_sixgrams(Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams)
  Sevengrams_metrics = cohesion_metrics_sevengrams(Unigrams, Bigrams, Threegrams, Fourgrams, Fivegrams, Sixgrams, Sevengrams)
  list_all_metrics = [Bigrams_metrics, Threegrams_metrics, Fourgrams_metrics, Fivegrams_metrics, Sixgrams_metrics, Sevengrams_metrics]

  #Remove expressions with low cohesion
  for i in range(len(list_all_metrics)):
    mean_scp, mean_dice, mean_mi, mean_omega = mean_for_all(list_all_metrics[i])
    list_all_metrics[i] = min_cohesion_all(list_all_metrics[i], mean_scp, mean_dice, mean_mi, mean_omega)

  #Remove expressions with puctuntation and frequency inferior to 2
  list_all_metrics = remove_infreq_punct(list_all_metrics)

  #Remove stop words
  if len(list_all_metrics[0]) > 0:
    retrieved_expressions = remove_stop_words(list_all_metrics, stop_words)
  else:
    retrieved_expressions = []

  return retrieved_expressions
```

```python
def prob(word , doc_index): 
  if word not in RE_docs_dict_50[doc_index]:
    return 0
  return RE_docs_dict_50[doc_index][word] / len(docs[doc_index])

  
def mean_prob(word):
  counter = 0
  for doc_index in range(len(docs)):
    counter += prob(word, doc_index)
  return counter / len(docs)


def cov(word1, word2):
  cov_sum = 0
  for doc_index in range(len(docs)):
    cov_sum = cov_sum + ( (prob(word1, doc_index) - mean_prob(word1)) * 
                        (prob(word2, doc_index) - mean_prob(word2)) )
  return cov_sum / (len(docs)-1) 


def correlation(word1, word2):
  return cov(word1, word2) / (math.sqrt(cov(word1, word1)) * math.sqrt(cov(word2, word2)))
```
